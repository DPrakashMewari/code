```python
df = sample.copy() # Make a copy of a data
```

```python
# function to remove backslash , And Percent total
def clean_marks_cloumn(x):
  
    [y1,y2]=x.split('/')
    
    return((np.float(y1)/np.float(y2))*100)
```

```python
# Group of Data science Profiles
df['DS_Profile']=df['Other skills'].str.contains('Machine Learning') | df['Other skills'].str.contains('Data Science')| df['Other skills'].str.contains('Natural Language Processing (NLP)') | df['Other skills'].str.contains('Data Analytics') | df['Other skills'].str.contains('Deep Learning') | df['Other skills'].str.contains('Big data') | df['Other skills'].str.contains('R Programming')   
df['DS_Profile']=df['DS_Profile'].map({True:1,False:0})
```

```python
# Group of Web Devlopement Profiles
df['WD_Profile']=df['Other skills'].str.contains('HTML') | df['Other skills'].str.contains('CSS') |  df['Other skills'].str.contains('JavaScript') | df['Other skills'].str.contains('AJAX') | df['Other skills'].str.contains('Bootstrap') | df['Other skills'].str.contains('MongoDB') | df['Other skills'].str.contains('reactjs') | df['Other skills'].str.contains('nodejs')  
df['WD_Profile']=df['WD_Profile'].map({True:2,False:0}) 
```

```python
df.describe() # max min,mean 
```

```python
df['Performance_PG']=df[['Performance_PG']].dropna().applymap(clean_marks_cloumn)
df['Performance_PG'].fillna(value=0,inplace=True)
```

```python
df[['Performance_10','Performance_12','Performance_UG']] = df[['Performance_10','Performance_12','Performance_UG']].dropna().applymap(clean_marks_cloumn)
df[['Performance_10','Performance_12','Performance_UG']].fillna(value=0,inplace=True)
```

```python
df=df.fillna(0)
```

```python
df.columns
```

```python
df["Profile"]=df["DS_Profile"]+df["WD_Profile"]
df["Profile"]=df["Profile"].map({1:"Data Scientist",2:"Web Developement",3:"Considered for both Data Scientist and Web Developement"})
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_values=StandardScaler()
df.iloc[:,2:15]=sc_values.fit_transform(df.iloc[:,2:15])


sc_marks=StandardScaler()
df.iloc[:,19:23]=sc_marks.fit_transform(df.iloc[:,19:23])
```

```python
df["Scores"]=0
```

```python
for index,rows in df.iterrows():
    score=0
    if(rows[1]=='Bangalore' or rows[20]=='2019'):
        
        score=10+rows[2]+rows[3]+rows[4]+rows[5]+rows[6]+rows[7]+rows[8]+rows[9]+rows[10]+rows[11]+rows[12]+rows[13]+rows[14]+rows[19]+10+rows[21]+rows[22]
       
    else:
        score=rows[2]+rows[3]+rows[4]+rows[5]+rows[6]+rows[7]+rows[8]+rows[9]+rows[10]+rows[11]+rows[12]+rows[13]+rows[14]+rows[19]+rows[20]+rows[21]+rows[22]
       
    df.iloc[index,-1] =score  
```

```python
df["Profile"]=df["DS_Profile"]+df["WD_Profile"]
df["Profile"]=df["Profile"].map({1:"Data Scientist",2:"Web Developement",3:"Considered for both Data Scientist and Web Developement"})
out_df=df[["Application_ID","Profile","Scores"]]
out_df.sort_values("Scores",ascending=False,inplace=True)    
```
